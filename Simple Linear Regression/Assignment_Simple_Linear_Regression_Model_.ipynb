## ASSIGNMENT SIMPLE LINEAR REGRESSION 


Q1 . Delivery_time -> Predict delivery time using sorting time 

```python
## 1st Import all necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df1 = pd.read_csv("/content/delivery_time.csv")
df1
```

```python
## 1st remove space and add underscore in column names
df1 = df1.rename(columns={"Delivery Time":"Del_Time","Sorting Time":"Sort_Time"})
df1 
```

```python
new_df1 = df1.drop(["Del_Time"],axis=1)
new_df1
```

```python
X_train,X_test,y_train,y_test = train_test_split(df1[['Del_Time']], df1['Sort_Time'], test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
Delivery_time = [[13.75]]
prediction = model.predict(Delivery_time)
print("Model Accuracy Score is : ",model.score(X_train,y_train)*100.0,"%")
print("Your prediction value is :",prediction)
```
