# Problem Statement
Perform Clustering for the crime data and identify the number of clusters formed and draw inferences.

Data Description:

Murder -- Muder rates in different places of United States

Assualt- Assualt rate in different places of United States

UrbanPop - urban population in different places of United States

Rape - Rape rate in different places of United States

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/crime_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(["Unnamed: 0"],axis = 1,inplace = True)
df.head()
```

```python
df_norm = StandardScaler().fit_transform(df)
df_norm
```

# DBSCAN

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(df_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
df['clusters']=dbscan.labels_
df
```

```python
df.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))
plt.scatter(df['clusters'],df['UrbanPop'], c=dbscan.labels_)
```

# K-Means
1.Use Elbow Graph to find optimum number of  clusters (K value) from K values range

2.The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS

3.Random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.

```python
# within-cluster sum-of-squares criterion

from sklearn.cluster import KMeans

wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=4

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(df_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
df1=df.copy()
df1['clusters4id']=clusters4.labels_
df1.head()
```

```python
# Compute the centroids for K=4 clusters with 11 variables
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
df1.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters for Murder
plt.figure(figsize=(10, 7))
plt.scatter(df1['clusters4id'],df1['Murder'], c=clusters4.labels_)
```

```python
# Plot Clusters for Assault
plt.figure(figsize=(10, 7))
plt.scatter(df1['clusters4id'],df1['Assault'], c=clusters4.labels_)
```

```python
# Plot Clusters for UrbanPop
plt.figure(figsize=(10, 7))
plt.scatter(df1['clusters4id'],df1['UrbanPop'], c=clusters4.labels_)
```

```python
# Plot Clusters for Rape
plt.figure(figsize=(10, 7))
plt.scatter(df1['clusters4id'],df1['Rape'], c=clusters4.labels_)
```
